```python

def run_tensorflow():
    import tensorflow as tf
    from tensorflow import keras
    from tensorflow.keras import layers
    import numpy as np

    print("=" * 60)
    print("TENSORFLOW / KERAS")
    print("=" * 60)
    print("TensorFlow Version:", tf.__version__)
    print("Keras Version:", keras.__version__)


    np.random.seed(42)
    num_samples = 1000
    num_features = 10

    X = np.random.rand(num_samples, num_features).astype(np.float32)
    y = (np.sum(X[:, :5], axis=1) > np.sum(X[:, 5:], axis=1)).astype(np.float32)

    print(f"Generated dataset: X.shape={X.shape}, y.shape={y.shape}")


    model = keras.Sequential([
        keras.Input(shape=(num_features,)),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(32, activation='relu'),
        layers.Dropout(0.2),
        layers.Dense(1, activation='sigmoid'),
    ])

    print("\nModel Summary:")
    model.summary()


    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy'],
    )
    print("\nModel compiled successfully.")


    epochs = 15
    batch_size = 32
    model.fit(X, y, epochs=epochs, batch_size=batch_size, verbose=1)
    print(f"\nModel training completed for {epochs} epochs.")

    loss, accuracy = model.evaluate(X, y, verbose=0)
    print(f"\nEvaluation — Loss: {loss:.4f}, Accuracy: {accuracy:.4f}")

    num_predictions = 5
    sample_indices = np.random.choice(num_samples, num_predictions, replace=False)
    X_new = X[sample_indices]
    y_true = y[sample_indices]
    predictions = model.predict(X_new)

    print(f"\nPredictions for {num_predictions} samples:")
    for i in range(num_predictions):
        prob = predictions[i][0]
        pred_cls = 1 if prob > 0.5 else 0
        true_cls = int(y_true[i])
        print(f"  Sample {i+1}: Prob={prob:.4f}, Predicted={pred_cls}, True={true_cls}")



def run_pytorch():
    import torch
    import torch.nn as nn
    import torch.optim as optim
    from torch.utils.data import TensorDataset, DataLoader
    import numpy as np

    print("\n" + "=" * 60)
    print("PYTORCH")
    print("=" * 60)
    print("PyTorch Version:", torch.__version__)


    np.random.seed(42)
    num_samples = 1000
    num_features = 10

    X_np = np.random.rand(num_samples, num_features).astype(np.float32)
    y_np = (np.sum(X_np[:, :5], axis=1) > np.sum(X_np[:, 5:], axis=1)).astype(np.float32)

    X = torch.from_numpy(X_np)
    y = torch.from_numpy(y_np).unsqueeze(1)

    print(f"Generated dataset: X.shape={X.shape}, y.shape={y.shape}")


    class BinaryClassifier(nn.Module):
        def __init__(self, input_size):
            super().__init__()
            self.net = nn.Sequential(
                nn.Linear(input_size, 64),
                nn.ReLU(),
                nn.Dropout(0.2),
                nn.Linear(64, 32),
                nn.ReLU(),
                nn.Dropout(0.2),
                nn.Linear(32, 1),
                nn.Sigmoid(),
            )

        def forward(self, x):
            return self.net(x)

    model = BinaryClassifier(num_features)
    print("\nModel Architecture:")
    print(model)


    criterion = nn.BCELoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)
    print("\nLoss function and optimizer defined successfully.")


    batch_size = 32
    dataloader = DataLoader(TensorDataset(X, y), batch_size=batch_size, shuffle=True)


    epochs = 15
    print(f"\nStarting training for {epochs} epochs...")
    for epoch in range(epochs):
        model.train()
        for inputs, labels in dataloader:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
    print("Model training completed.")


    model.eval()
    with torch.no_grad():
        outputs = model(X)
        loss = criterion(outputs, y)
        predicted_classes = (outputs > 0.5).float()
        accuracy = (predicted_classes == y).float().mean()
    print(f"\nEvaluation — Loss: {loss.item():.4f}, Accuracy: {accuracy.item():.4f}")

    num_predictions = 5
    sample_indices = np.random.choice(num_samples, num_predictions, replace=False)
    X_new = X[sample_indices]
    y_true = y_np[sample_indices]

    with torch.no_grad():
        predictions = model(X_new)

    print(f"\nPredictions for {num_predictions} samples:")
    for i in range(num_predictions):
        prob = predictions[i].item()
        pred_cls = 1 if prob > 0.5 else 0
        true_cls = int(y_true[i])
        print(f"  Sample {i+1}: Prob={prob:.4f}, Predicted={pred_cls}, True={true_cls}")



if __name__ == "__main__":
    run_tensorflow()
    run_pytorch()
```

```python
PART 1: TensorFlow / Keras
🔹 Dataset Generation
The code first creates a synthetic dataset using random numbers
Each sample has 10 features
The label is decided by a rule:
If the sum of first 5 features is greater than last 5, output = 1
Otherwise, output = 0

👉 So the model learns to compare two halves of input features

🔹 Model Architecture
A neural network is created using multiple layers
It includes:
Dense layers for learning patterns
Dropout layers to reduce overfitting
The final layer uses sigmoid activation to give output between 0 and 1

👉 Output represents probability of class 1

🔹 Model Compilation
The model is configured with:
Binary cross-entropy loss (for 2 classes)
Adam optimizer (for efficient learning)
Accuracy as performance metric
🔹 Model Training
The model is trained on the dataset for multiple iterations (epochs)
During training:
It learns the relationship between input features and labels
Adjusts weights to reduce prediction error
🔹 Model Evaluation
After training, the model is evaluated on the same dataset
It calculates:
Loss (error)
Accuracy (correct predictions)
🔹 Predictions
The model is tested on a few random samples
It outputs probabilities
These probabilities are converted into:
1 (if probability > 0.5)
0 otherwise

👉 Shows how well the model predicts new data

🔷 PART 2: PyTorch
🔹 Dataset Generation
Same dataset is created again using the same logic
Data is converted into PyTorch tensors

👉 PyTorch works with tensors instead of arrays

🔹 Model Architecture
A neural network is defined using layers:
Linear layers (like dense layers)
ReLU activation for non-linearity
Dropout for regularization
Final layer uses sigmoid for binary output

👉 Structure is same as TensorFlow but written differently

🔹 Loss and Optimizer
Uses:
Binary Cross Entropy Loss
Adam optimizer

👉 Same logic as TensorFlow, different syntax

🔹 Data Handling
Data is split into batches using a DataLoader
This helps in efficient training
🔹 Training Process
Training is done manually in a loop
For each batch:
Model predicts output
Loss is calculated
Gradients are computed
Weights are updated

👉 Unlike TensorFlow, everything is explicitly controlled

🔹 Evaluation
After training:
Model predictions are generated
Compared with actual labels
Accuracy is calculated manually
🔹 Predictions
Random samples are selected
Model predicts probability
Converted to class (0 or 1)
```
