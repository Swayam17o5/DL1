```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import mnist

from sklearn.metrics import precision_score, recall_score, f1_score

(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.reshape((x_train.shape[0], 28 * 28)).astype('float32') / 255.0
x_test = x_test.reshape((x_test.shape[0], 28 * 28)).astype('float32') / 255.0

model = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\nTraining the model...")
history = model.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)

print("\nEvaluating the model...")
loss, accuracy = model.evaluate(x_test, y_test)

print(f"Test Loss     : {loss:.4f}")
print(f"Test Accuracy : {accuracy:.4f}")

y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)

precision = precision_score(y_test, y_pred_classes, average='macro')
recall = recall_score(y_test, y_pred_classes, average='macro')
f1 = f1_score(y_test, y_pred_classes, average='macro')

print(f"Precision     : {precision:.4f}")
print(f"Recall        : {recall:.4f}")
print(f"F1 Score      : {f1:.4f}")

print("\nModel Summary:")
model.summary()

import tensorflow as tf
import numpy as np

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import fashion_mnist

from sklearn.metrics import precision_score, recall_score, f1_score

(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

x_train = x_train.reshape((x_train.shape[0], 28 * 28)).astype('float32') / 255.0
x_test = x_test.reshape((x_test.shape[0], 28 * 28)).astype('float32') / 255.0

model = Sequential([
    Dense(128, activation='relu', input_shape=(784,)),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\nTraining the model...")
history = model.fit(
    x_train,
    y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.2
)

print("\nEvaluating the model...")
loss, accuracy = model.evaluate(x_test, y_test)

print(f"Test Loss     : {loss:.4f}")
print(f"Test Accuracy : {accuracy:.4f}")

y_pred = model.predict(x_test)
y_pred_classes = np.argmax(y_pred, axis=1)

precision = precision_score(y_test, y_pred_classes, average='macro')
recall = recall_score(y_test, y_pred_classes, average='macro')
f1 = f1_score(y_test, y_pred_classes, average='macro')

print(f"Precision     : {precision:.4f}")
print(f"Recall        : {recall:.4f}")
print(f"F1 Score      : {f1:.4f}")

print("\nModel Summary:")
model.summary()
```

```python
PART 1: Handwritten Digit Classification (MNIST)
🔹 Dataset Loading
The code loads the MNIST dataset
Contains:
60,000 training images
10,000 test images
Each image:
Size = 28×28 pixels
Grayscale

👉 Labels represent digits 0–9

🔹 Data Preprocessing
Images are reshaped:
From 28×28 → 784 (flattened vector)
Pixel values normalized:
From 0–255 → 0–1

👉 Required for faster and stable training

🔹 Model Architecture
A fully connected (Dense) neural network is used
Layers:
Input Layer
784 neurons (one for each pixel)
Hidden Layer
128 neurons with ReLU activation
Learns patterns in data
Output Layer
10 neurons (digits 0–9)
Softmax activation → gives probabilities

👉 No convolution here (unlike CNN)

🔹 Model Compilation
Uses:
Adam optimizer
Sparse categorical crossentropy loss
Accuracy as metric

👉 Labels are integers (not one-hot)

🔹 Model Training
Trained for 10 epochs
Batch size = 32
20% data used for validation

👉 Model learns digit patterns from pixels

🔹 Model Evaluation
Tested on unseen data
Outputs:
Test loss
Test accuracy
🔹 Predictions
Model predicts probabilities for each class
Highest probability → predicted digit
🔹 Performance Metrics
Precision
How many predicted values are correct
Recall
How many actual values are correctly predicted
F1 Score
Balance between precision and recall

👉 Calculated using sklearn

🔹 Model Summary
Displays:
Layers
Number of parameters
🔷 PART 2: Fashion Image Classification (Fashion MNIST)
🔹 Dataset Loading
The code loads the Fashion MNIST dataset
Contains images of:
Shirts, shoes, bags, etc.

👉 Same structure as MNIST but more complex

🔹 Data Preprocessing
Same steps:
Flatten images (28×28 → 784)
Normalize pixel values
🔹 Model Architecture
Same neural network:
Dense (128) → ReLU
Dense (10) → Softmax

👉 No change in architecture

🔹 Model Training
Trained for 20 epochs (more than MNIST)
Because:
Fashion dataset is harder
🔹 Model Evaluation
Outputs:
Test accuracy
Test loss

👉 Usually lower accuracy than MNIST

🔹 Predictions & Metrics
Same process:
Predict classes
Calculate precision, recall, F1-score
```
