McCulloch-Pitts (AND GATE)

```python
# McCulloch-Pitts Neuron for AND Gate (Truth Table)

def step_function(net, threshold):
    if net >= threshold:
        return 1
    else:
        return 0

# Weights and threshold for AND gate
w1 = 1
w2 =-1
threshold = 1.5

print("AND Gate using McCulloch-Pitts Model\n")
print("x1   x2   net   output")
print("----------------------")

# Generate all combinations
for x1 in [0, 1]:
    for x2 in [0, 1]:
        net = (x1 * w1) + (x2 * w2)
        output = step_function(net, threshold)
        print(f"{x1}    {x2}    {net}     {output}")
```

McCulloch-Pitts (USE CASE)

```python
# McCulloch-Pitts Loan Approval System

def step_function(net, threshold):
    if net >= threshold:
        return 1
    else:
        return 0

# Weights
w1 = 1   # Credit Score
w2 = 1   # Income Stability
w3 = 1   # Low Debt

threshold = 2

print("Loan Approval System using McCulloch-Pitts Model\n")
print("x1  x2  x3  net  decision")
print("-------------------------")

for x1 in [0, 1]:
    for x2 in [0, 1]:
        for x3 in [0, 1]:
            net = (x1*w1) + (x2*w2) + (x3*w3)
            output = step_function(net, threshold)
            decision = "Approve" if output == 1 else "Reject"
            print(f"{x1}   {x2}   {x3}   {net}    {decision}")
```

Perceptron Model(AND GATE)

```python
# AND Gate using Perceptron Learning Algorithm

# Training Data for AND gate
inputs = [(0,0), (0,1), (1,0), (1,1)]
targets = [0, 0, 0, 1]

# Initialize weights and bias
w1 = -1
w2 = 1
bias = 0
learning_rate = 0.1

# Activation Function
def step(net):
    return 1 if net >= 0 else 0

# Training
for epoch in range(10):
    for i in range(len(inputs)):
        x1, x2 = inputs[i]
        target = targets[i]

        net = x1*w1 + x2*w2 + bias
        output = step(net)

        error = target - output

        # Update weights
        w1 = w1 + learning_rate * error * x1
        w2 = w2 + learning_rate * error * x2
        bias = bias + learning_rate * error

print("Trained Weights:")
print("w1 =", w1, " w2 =", w2, " bias =", bias)

# Testing
print("\nAND Gate Results:")
print("x1  x2  output")
print("--------------")

for x1, x2 in inputs:
    net = x1*w1 + x2*w2 + bias
    y = step(net)
    print(f"{x1}   {x2}   {y}")
```

Perceptron (OR GATE)

```python
# OR Gate using Perceptron Learning Algorithm

# Training Data for OR gate
inputs = [(0,0), (0,1), (1,0), (1,1)]
targets = [0, 1, 1, 1]   # OR gate outputs

# Initialize weights and bias
w1 = 0
w2 = 1
bias = 0
learning_rate = 0.1

# Activation Function
def step(net):
    return 1 if net >= 0 else 0

# Training
for epoch in range(10):
    for i in range(len(inputs)):
        x1, x2 = inputs[i]
        target = targets[i]

        net = x1*w1 + x2*w2 + bias
        output = step(net)

        error = target - output

        # Update weights
        w1 = w1 + learning_rate * error * x1
        w2 = w2 + learning_rate * error * x2
        bias = bias + learning_rate * error

print("Trained Weights:")
print("w1 =", w1, " w2 =", w2, " bias =", bias)

# Testing
print("\nOR Gate Results:")
print("x1  x2  output")
print("--------------")

for x1, x2 in inputs:
    net = x1*w1 + x2*w2 + bias
    y = step(net)
    print(f"{x1}   {x2}   {y}")
```

```python
import numpy as np

X = np.array([[0,0], [0,1], [1,0], [1,1]])
Y = np.array([[0], [1], [1], [0]])

sig = lambda x: 1 / (1 + np.exp(-x))
dsig = lambda x: x * (1 - x)

np.random.seed(1)

w1 = np.random.rand(2, 3)
w2 = np.random.rand(3, 1)

lr = 0.5

for i in range(5000):

    # Forward pass
    h = sig(X @ w1)
    o = sig(h @ w2)

    # Error
    err = Y - o

    # Backpropagation
    w2 += lr * h.T @ (err * dsig(o))
    w1 += lr * X.T @ ((err * dsig(o)) @ w2.T * dsig(h))

    # Print at epoch 1 and 10
    if i == 0:
        print("Epoch 1")
        print("w1 =\n", w1)
        print("w2 =\n", w2)

    if i == 9:
        print("\nEpoch 10")
        print("w1 =\n", w1)
        print("w2 =\n", w2)

print("\nFinal Output:")
print(np.round(o, 3))
```
