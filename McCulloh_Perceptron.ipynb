```python
def step_function(net, threshold):
    return 1 if net >= threshold else 0

w1, w2, w3 = 1, 1, 1
threshold = 2

print("Loan Approval System using McCulloch-Pitts Model\n")
print("x1  x2  x3  net  decision")
print("-------------------------")

for x1 in [0, 1]:
    for x2 in [0, 1]:
        for x3 in [0, 1]:
            net = (x1*w1) + (x2*w2) + (x3*w3)
            output = step_function(net, threshold)
            decision = "Approve" if output == 1 else "Reject"
            print(f"{x1}   {x2}   {x3}   {net}    {decision}")
```

```python
inputs = [(0,0), (0,1), (1,0), (1,1)]
targets = [0, 0, 0, 1]

w1, w2, bias = 0, 0, 0
lr = 0.1

def step(net):
    return 1 if net >= 0 else 0

for epoch in range(10):
    for i in range(len(inputs)):
        x1, x2 = inputs[i]
        target = targets[i]

        net = x1*w1 + x2*w2 + bias
        output = step(net)

        error = target - output

        w1 += lr * error * x1
        w2 += lr * error * x2
        bias += lr * error

print("Trained Weights:")
print("w1 =", w1, " w2 =", w2, " bias =", bias)

print("\nAND Gate Results:")
print("x1  x2  output")
print("--------------")

for x1, x2 in inputs:
    net = x1*w1 + x2*w2 + bias
    print(f"{x1}   {x2}   {step(net)}")
```
